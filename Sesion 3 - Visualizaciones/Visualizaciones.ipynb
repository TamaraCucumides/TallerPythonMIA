# Personalizacion de graficos

```python
## colores: hay colores predeterminados

plt.plot(x, color="r")
plt.show()
```

```python
## O podemos identificarlos por su codigo HEX

plt.plot(x, color="#D6D2FF")
plt.show()
```

```python
## O varios colores

colores = ["#d0e7b5", "#d6d2ff", "#ffb4fc", "#a8daff", "#bffaff"]


plt.bar(x,y, color=colores)
```

```python
## annotate: podemos escribir cosas en cualquier lugar de nuestro grafico

fig = plt.plot(x)
plt.annotate('Una anotacion', xy =(5, 2))
```

```python
## subplots: podemos hacer multiples graficos en una misma figura

fig, axs = plt.subplots(2)
fig.suptitle('Dos subplots alineados verticalmente')
axs[0].plot(x, y) ##el primer grafico
axs[1].plot(x, -y) ##el segundo grafico
plt.show()
```

```python
## tambien horizontalmente

fig, (ax1, ax2) = plt.subplots(1, 2) ## este es nuestro shape, 1 fila, 2 columnas
fig.suptitle('Ahora dos plots alineados horizontalmente')
ax1.plot(x, y)
ax2.plot(x, -y)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 2) ## este es nuestro shape, 1 fila, 2 columnas
fig.suptitle('Ahora cuatro axes en una misma figura')
ax1[0].plot(x, y)
ax2[0].plot(x, -y)
ax1[1].plot(y, x)
ax2[1].plot(y, -x)
plt.show()
```

```python
## Cada subplot puede ser distinto y podemos cambiar sus colores, leyendas, etc.

fig, (ax1, ax2) = plt.subplots(2, 2) ## este es nuestro shape, 1 fila, 2 columnas
fig.suptitle('Ahora cuatro plots')
ax1[0].plot(x, y, color="r")
ax2[0].bar(x, y, color=colores)
ax1[1].pie(y)
ax2[1].scatter(y, -x, marker="x")
plt.show()
```

```python
## lo de arriba es muy pequeo, agrandemoslo

plt.rcParams["figure.figsize"] = (8,8)

fig, (ax1, ax2) = plt.subplots(2, 2) ## este es nuestro shape, 1 fila, 2 columnas
fig.suptitle('Ahora cuatro plots')
ax1[0].plot(x, y, color="r")
ax2[0].bar(x, y, color=colores)
ax1[1].pie(y)
ax2[1].scatter(y, -x, marker="x")
plt.show()
```

```python
## si queremos setear un porte para todos los graficos podemos hacer
plt.rcParams["figure.figsize"] = (8,8)
```

```python
## Cada plot tiene sus hiperparametros que nos permiten cambiar cosas

plt.scatter(x,y, marker="x")
plt.show()
```

# Ms opciones para grficas

Veamos alguns grficas mas personalizadas. 

## Dataset tips

```python
tips = sns.load_dataset("tips")
```

```python
type(tips)
```

```python
tips
```

```python
## Boxplots por genero

sns.boxplot(data=tips, x="tip", y="sex")
plt.show()
```

```python
## una alternativa a boxplot son los swarmplot o los violinplot

sns.violinplot(data=tips, x="tip", y="sex")
plt.show()
```

```python
## Podemos mezclar graficos

sns.boxplot(data=tips, x="tip", y="sex")
sns.swarmplot(data=tips, x="tip", y="sex")
plt.show()
```

```python
sns.barplot(data=tips, x="smoker", y="tip", hue="sex")
plt.show()
```

```python
## Seaborn trae algunas paletas bonitas predefinidas (tambien podemos poner una paleta personalizada)

sns.barplot(data=tips, x="smoker", y="tip", hue="sex", palette="husl")
plt.show()
```

Podemos tener un heatmap que nos muestre las correlaciones entre las variables

```python
sns.heatmap(tips.corr(), annot=True)
plt.show()
```
